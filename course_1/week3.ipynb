## Activation Functions

* An activation function is a mathematical equation applied to the output of a neural network neuron. Its primary purpose is to introduce non-linearity. Without it, the network acts as a simple linear model, unable to learn the complex patterns required to solve tasks like image classification or natural language processing.
* Sigmoid -  Squashes the output into a continuous range between 0 and 1. Its steep slope is useful for gradients, but it is prone to the "vanishing gradient" problem during deep training.
* ReLU (Rectified Linear Unit) -  It outputs zero if the input is negative, and the exact value if the input is positive. It is computationally efficient and currently the industry standard.
* Leaky ReLU - Similar to ReLU, but allows a very small, non-zero gradient (e.g., $(0.01x)$) for negative inputs. This prevents "dead neurons" in situations with noisy data.
* Tanh -  Squashes values into a range between -1 and 1. Because the output is zero-centered, it is generally easier to train than the sigmoid function. Its a shifted version of sigmoid function.

<img src="./images/dl_13.png" width=400 height=300>

* Softmax - Used for output layer. Used for multi-class classification problems. It converts raw scores into probability distributions that sum to 1.
* The activation function can be different for different layers. For example tanh and then sigmoid for output to have output values between 0 and 1.
* When the pre-activation $(z)$ is large, both sigmoid and tanh functions experience saturation. This flattens their curves and drives their derivatives toward zero, resulting in the notorious vanishing gradient problem.
* If you are doing binary classification use sigmoid for output layer. If you are uncertain what to use for the rest then use ReLU.
* The identity activation function is just the linear portion without any non-linear transformation. Nonlinear activation functions are what make neural networks powerful. Without them, even very deep networks collapse into a simple linear model and cannot learn complex patterns in data. Nonlinearities such as ReLU, tanh, and sigmoid enable neural networks to represent much more sophisticated functions.

# Backpropagation: An Intuitive Tutorial

Most explanations of backpropagation start with equations. I think it's easier to start with a question:

> **If a neural network makes a mistake, how do we figure out which weight caused it?**

Backpropagation is the algorithm that answers that question.

---

# Step 1: Think Like a Student Taking a Test

Imagine a student takes an exam.

* Correct answer: 95
* Student answer: 70

The student is off by 25 points.

Now suppose the student got that answer through several intermediate calculations:

```text
Input -> Step A -> Step B -> Final Answer
```

To improve next time, we need to know:

* How much did Step B contribute to the error?
* How much did Step A contribute?
* What should change?

Backpropagation does exactly this for neural networks.

---

# Step 2: A Neural Network is Just a Chain of Calculations

Consider the simplest neural network:

```text
x
 |
 v
z = wx + b
 |
 v
a = sigmoid(z)
 |
 v
Loss
```

Forward propagation means:

1. Compute z
2. Compute a
3. Compute loss

Example:

```text
x = 2
w = 3
b = 1

z = 3*2 + 1 = 7

a = sigmoid(7) ≈ 0.999
```

Suppose the true label is:

```text
y = 0
```

Then our prediction is very wrong.

Now we need to know:

> Which parameter should change, and by how much?

---

# Step 3: The Core Idea of Backpropagation

Start at the error.

Ask:

> If I slightly changed z, would the loss get better or worse?

That's:

$$
\frac{\partial L}{\partial z}
$$

Then ask:

> If I slightly changed w, how would z change?

That's:

$$
\frac{\partial z}{\partial w}
$$

Combine them:

$$
\frac{\partial L}{\partial w}
=

\frac{\partial L}{\partial z}
\cdot
\frac{\partial z}{\partial w}
$$

This is just the **chain rule**.

Backpropagation is basically:

> Repeatedly applying the chain rule from the output backwards.

---

# Step 4: Why It's Called "Back" Propagation

Forward pass:

```text
x
 ↓
z
 ↓
a
 ↓
Loss
```

Backward pass:

```text
Loss
 ↑
a
 ↑
z
 ↑
w,b
```

The error signal travels backward through the network.

Hence:

**Back-propagation**

---

# Step 5: Logistic Regression Example

For logistic regression:

```text
x
 |
 v
z
 |
 v
sigmoid
 |
 v
a
 |
 v
Loss
```

A beautiful result appears:

$$
dz = a - y
$$

This means:

```text
Error = Prediction - Truth
```

Example:

```text
Prediction = 0.9
Truth      = 1
```

Then

```text
dz = -0.1
```

Network says:

> "Increase the output slightly."

If:

```text
Prediction = 0.9
Truth      = 0
```

Then

```text
dz = 0.9
```

Network says:

> "Decrease the output a lot."

This is why

$$
dz = a-y
$$

is so elegant.

---

# Step 6: Computing Weight Gradients

Suppose:

```text
x = 5
dz = 0.2
```

Then:

$$
dw = dz \cdot x
$$

So:

```text
dw = 0.2 * 5 = 1
```

Interpretation:

The large input value contributed strongly to the error.

The gradient becomes larger.

---

# Step 7: Add a Hidden Layer

Now we have:

```text
Input
  |
  v
Hidden Layer
  |
  v
Output Layer
```

Or:

```text
x
 |
 v
z1
 |
 v
a1
 |
 v
z2
 |
 v
a2
 |
 v
Loss
```

Forward pass:

```text
x
→ z1
→ a1
→ z2
→ a2
→ Loss
```

Backpropagation:

```text
Loss
→ dz2
→ dw2
→ db2
→ dz1
→ dw1
→ db1
```

The same pattern repeats.

---

# Step 8: What Happens at the Output Layer?

Output layer error:

$$
dz^{[2]}
=

a^{[2]} - y
$$

Then:

$$
dw^{[2]}
=

dz^{[2]}(a^{[1]})^T
$$

Notice:

```text
Output error
×
Input into that layer
```

This pattern appears everywhere.

---

# Step 9: How Error Reaches the Hidden Layer

This is the most important conceptual step.

The hidden layer doesn't know the correct answer.

Only the output layer does.

So how can a hidden neuron know it made a mistake?

The answer:

The output layer sends error information backward.

Mathematically:

$$
dz^{[1]}
=

(W^{[2]})^T dz^{[2]}
\odot
g'(z^{[1]})
$$

Interpretation:

### Part 1

$$
(W^{[2]})^T dz^{[2]}
$$

means:

> "How much of the output error belongs to each hidden neuron?"

### Part 2

$$
g'(z^{[1]})
$$

means:

> "How sensitive is this hidden neuron?"

Multiply them together:

```text
Responsibility
×
Sensitivity
=
Hidden-layer error
```

That's the intuition.

---

They are shape mistakes.

For example:

```text
(3×4)
times
(4×1)

works
```

But:

```text
(3×4)
times
(5×1)

fails
```

Whenever you write:

$$

$$

always check:

```text
columns(A) = rows(B)
```

Many professional ML engineers still use dimension checking as their primary debugging technique.

## Random Initialization

### Summary: Why Neural Networks Need Random Weight Initialization

This segment explains why **initializing all neural network weights to zero fails** and why **small random initialization** is necessary.

#### 1. Zero Initialization Causes a Symmetry Problem

* Consider a neural network with:

  * 2 input features
  * 2 hidden units
* If all weights ($W^{[1]}$) are initialized to zero and biases ($b^{[1]}$) are also zero:

  * Both hidden units compute exactly the same function.
  * Their activations are identical.
  * During backpropagation, their gradients are also identical.
  * After each gradient descent update, the weights remain identical.

As a result:

* The hidden units remain **symmetric** throughout training.
* They never learn different features.
* Having multiple hidden units becomes equivalent to having just one hidden unit.

This problem extends to larger networks: if all weights start at zero, all neurons in the same layer remain identical and learn the same thing.

---

#### 2. Random Initialization Breaks Symmetry

To solve this, initialize weights with small random values:

```python
W1 = np.random.randn(2,2) * 0.01
b1 = np.zeros((2,1))

W2 = np.random.randn(1,2) * 0.01
b2 = np.zeros((1,1))
```

Key idea:

* Different random weights make neurons start with different computations.
* This breaks symmetry.
* Hidden units can learn different features during training.

---

#### 3. Why Small Random Values?

Using very large random weights is also problematic.

Since:
$$
[
Z = WX + b
]
$$
large weights produce large (Z) values.

For sigmoid or tanh activations:

* Large positive or negative (Z) values push activations into the saturated regions.
* Saturated regions have very small gradients.
* Small gradients slow down learning significantly.

Therefore:

* Initialize weights randomly **and** keep them small (e.g., multiply by 0.01).

---

#### 4. Biases Can Be Initialized to Zero

Unlike weights:

* Biases do not create the symmetry problem.
* It is generally safe to initialize all bias terms to zero.

---

#### 5. Deep Networks May Need Better Initialization

* For shallow networks (e.g., one hidden layer), multiplying by 0.01 often works well.
* For very deep networks, more sophisticated initialization methods may perform better.
* These methods choose different scaling constants to improve training speed and stability.

---

### Main Takeaways

* **Never initialize all weights to zero in a neural network.**
* Zero weights cause neurons in the same layer to remain identical and learn the same features.
* **Initialize weights with small random values** to break symmetry.
* **Biases can safely be initialized to zero.**
* Keeping initial weights small helps avoid activation saturation and slow learning.

# 16. Main Idea

Using this convention, weights are stored as columns.

That is why forward propagation uses transposes:

$$
Z^{[1]} = (W^{[1]})^T X + b^{[1]}
$$

$$
Z^{[2]} = (W^{[2]})^T A^{[1]} + b^{[2]}
$$

Backpropagation moves in the reverse direction:

$$
dZ^{[2]} \rightarrow dW^{[2]}, db^{[2]} \rightarrow dA^{[1]} \rightarrow dZ^{[1]} \rightarrow dW^{[1]}, db^{[1]}
$$

The averages happen in the gradient calculations:

$$
dW^{[2]} = \frac{1}{m} A^{[1]}(dZ^{[2]})^T
$$

$$
db^{[2]} = \frac{1}{m} \sum dZ^{[2]}
$$

$$
dW^{[1]} = \frac{1}{m} X(dZ^{[1]})^T
$$

$$
db^{[1]} = \frac{1}{m} \sum dZ^{[1]}
$$

So during one batch or epoch, each training example contributes to the gradients, then those gradient contributions are averaged before the weights and biases are updated.

# Derivations of the Previous:

$$
3 \rightarrow 4 \rightarrow 1
$$

we have:

$$
X \in \mathbb{R}^{3 \times m}
$$

$$
W^{[1]} \in \mathbb{R}^{3 \times 4}
$$

$$
b^{[1]} \in \mathbb{R}^{4 \times 1}
$$

$$
W^{[2]} \in \mathbb{R}^{4 \times 1}
$$

$$
b^{[2]} \in \mathbb{R}^{1 \times 1}
$$

Using the transpose convention:

$$
Z^{[1]} = (W^{[1]})^T X + b^{[1]}
$$

$$
A^{[1]} = g^{[1]}(Z^{[1]})
$$

$$
Z^{[2]} = (W^{[2]})^T A^{[1]} + b^{[2]}
$$

$$
A^{[2]} = \sigma(Z^{[2]})
$$

For binary classification:

$$
Y \in \mathbb{R}^{1 \times m}
$$

and the loss is binary cross-entropy:

$$
J = -\frac{1}{m}\sum_{i=1}^{m}
\left[
y^{(i)}\log(a^{[2](i)}) + (1-y^{(i)})\log(1-a^{[2](i)})
\right]
$$

---

# 1. Output Layer Backpropagation

We start at the output.

For sigmoid output with binary cross-entropy loss, the derivative simplifies nicely:

$$
dZ^{[2]} = A^{[2]} - Y
$$

Shape:

$$
dZ^{[2]} \in \mathbb{R}^{1 \times m}
$$

This says: prediction minus actual label.

---

# 2. Derive $dW^{[2]}$

The output score is:

$$
Z^{[2]} = (W^{[2]})^T A^{[1]} + b^{[2]}
$$

For a single example:

$$
z^{[2](i)} = (W^{[2]})^T a^{[1](i)} + b^{[2]}
$$

Expanded:

$$
z^{[2](i)} =
w_1^{[2]}a_1^{[1](i)}
+
w_2^{[2]}a_2^{[1](i)}
+
w_3^{[2]}a_3^{[1](i)}
+
w_4^{[2]}a_4^{[1](i)}
+
b^{[2]}
$$

For one output weight:

$$
\frac{\partial z^{[2](i)}}{\partial w_j^{[2]}}
=

a_j^{[1](i)}
$$

Using the chain rule:

$$
\frac{\partial J}{\partial w_j^{[2]}}
=

\frac{1}{m}
\sum_{i=1}^{m}
\frac{\partial J}{\partial z^{[2](i)}}
\frac{\partial z^{[2](i)}}{\partial w_j^{[2]}}
$$

Since:

$$
\frac{\partial J}{\partial z^{[2](i)}} = dz^{[2](i)}
$$

we get:

$$
\frac{\partial J}{\partial w_j^{[2]}}
=

\frac{1}{m}
\sum_{i=1}^{m}
dz^{[2](i)}a_j^{[1](i)}
$$

Stacking all 4 output weights into a vector gives:

$$
dW^{[2]} =
\frac{1}{m}A^{[1]}(dZ^{[2]})^T
$$

Shape check:

$$
A^{[1]} \in \mathbb{R}^{4 \times m}
$$

$$
(dZ^{[2]})^T \in \mathbb{R}^{m \times 1}
$$

Therefore:

$$
dW^{[2]} \in \mathbb{R}^{4 \times 1}
$$

which matches:

$$
W^{[2]} \in \mathbb{R}^{4 \times 1}
$$

---

# 3. Derive $db^{[2]}$

For each example:

$$
z^{[2](i)} = (W^{[2]})^T a^{[1](i)} + b^{[2]}
$$

So:

$$
\frac{\partial z^{[2](i)}}{\partial b^{[2]}} = 1
$$

Therefore:

$$
db^{[2]}
=

\frac{1}{m}
\sum_{i=1}^{m}
dz^{[2](i)}
$$

In vectorized form:

$$
db^{[2]} =
\frac{1}{m}\sum dZ^{[2]}
$$

Shape:

$$
db^{[2]} \in \mathbb{R}^{1 \times 1}
$$

---

# 4. Send the Error Back to the Hidden Layer

Now we need to figure out how much the output-layer error depends on each hidden activation.

Since:

$$
Z^{[2]} = (W^{[2]})^T A^{[1]} + b^{[2]}
$$

for a single example:

$$
z^{[2](i)}
=

w_1^{[2]}a_1^{[1](i)}
+
w_2^{[2]}a_2^{[1](i)}
+
w_3^{[2]}a_3^{[1](i)}
+
w_4^{[2]}a_4^{[1](i)}
+
b^{[2]}
$$

So:

$$
\frac{\partial z^{[2](i)}}{\partial a_j^{[1](i)}}
=

w_j^{[2]}
$$

By the chain rule:

$$
dA^{[1]} = W^{[2]}dZ^{[2]}
$$

Shape check:

$$
W^{[2]} \in \mathbb{R}^{4 \times 1}
$$

$$
dZ^{[2]} \in \mathbb{R}^{1 \times m}
$$

Therefore:

$$
dA^{[1]} \in \mathbb{R}^{4 \times m}
$$

This tells us how much the loss changes with respect to each hidden activation.

---

# 5. Convert $dA^{[1]}$ into $dZ^{[1]}$

The hidden activations came from:

$$
A^{[1]} = g^{[1]}(Z^{[1]})
$$

So we apply the activation derivative:

$$
dZ^{[1]} = dA^{[1]} \odot g^{[1]\prime}(Z^{[1]})
$$

If the hidden layer uses sigmoid:

$$
g^{[1]}(z) = \sigma(z)
$$

then:

$$
g^{[1]\prime}(z) = \sigma(z)(1-\sigma(z))
$$

Since:

$$
A^{[1]} = \sigma(Z^{[1]})
$$

we can write:

$$
dZ^{[1]} = dA^{[1]} \odot A^{[1]}(1-A^{[1]})
$$

Shape:

$$
dZ^{[1]} \in \mathbb{R}^{4 \times m}
$$

If the hidden layer uses ReLU instead:

$$
g^{[1]}(z)=\max(0,z)
$$

then:

$$
dZ^{[1]} = dA^{[1]} \odot \mathbf{1}(Z^{[1]} > 0)
$$

---

# 6. Derive $dW^{[1]}$

The hidden layer score is:

$$
Z^{[1]} = (W^{[1]})^T X + b^{[1]}
$$

For one example:

$$
z^{[1](i)} = (W^{[1]})^T x^{(i)} + b^{[1]}
$$

For hidden neuron (j):

$$
z_j^{[1](i)}
=

(w_j^{[1]})^T x^{(i)} + b_j^{[1]}
$$

Expanded:

$$
z_j^{[1](i)}
=

w_{1j}^{[1]}x_1^{(i)}
+
w_{2j}^{[1]}x_2^{(i)}
+
w_{3j}^{[1]}x_3^{(i)}
+
b_j^{[1]}
$$

For one hidden-layer weight:

$$
\frac{\partial z_j^{[1](i)}}{\partial w_{kj}^{[1]}}
=

x_k^{(i)}
$$

By the chain rule:

$$
\frac{\partial J}{\partial w_{kj}^{[1]}}
=

\frac{1}{m}
\sum_{i=1}^{m}
dz_j^{[1](i)}x_k^{(i)}
$$

Vectorizing over all input features and hidden neurons:

$$
dW^{[1]} = \frac{1}{m}X(dZ^{[1]})^T
$$

Shape check:

$$
X \in \mathbb{R}^{3 \times m}
$$

$$
(dZ^{[1]})^T \in \mathbb{R}^{m \times 4}
$$

Therefore:

$$
dW^{[1]} \in \mathbb{R}^{3 \times 4}
$$

which matches:

$$
W^{[1]} \in \mathbb{R}^{3 \times 4}
$$

---

# 7. Derive $db^{[1]}$

For hidden neuron (j):

$$
z_j^{[1](i)}
=

(w_j^{[1]})^T x^{(i)} + b_j^{[1]}
$$

So:

$$
\frac{\partial z_j^{[1](i)}}{\partial b_j^{[1]}} = 1
$$

Therefore:

$$
db_j^{[1]}
=

\frac{1}{m}
\sum_{i=1}^{m}
dz_j^{[1](i)}
$$

Vectorized:

$$
db^{[1]} =
\frac{1}{m}
\sum_{\text{columns}} dZ^{[1]}
$$

Shape:

$$
db^{[1]} \in \mathbb{R}^{4 \times 1}
$$

---

# Final Backpropagation Equations

For this network:

$$
3 \rightarrow 4 \rightarrow 1
$$

with:

$$
Z^{[1]} = (W^{[1]})^T X + b^{[1]}
$$

and:

$$
Z^{[2]} = (W^{[2]})^T A^{[1]} + b^{[2]}
$$

the backpropagation equations are:

$$
dZ^{[2]} = A^{[2]} - Y
$$

$$
dW^{[2]} = \frac{1}{m}A^{[1]}(dZ^{[2]})^T
$$

$$
db^{[2]} = \frac{1}{m}\sum_{\text{columns}} dZ^{[2]}
$$

$$
dA^{[1]} = W^{[2]}dZ^{[2]}
$$

$$
dZ^{[1]} = dA^{[1]} \odot g^{[1]\prime}(Z^{[1]})
$$

For sigmoid hidden activation:

$$
dZ^{[1]} = dA^{[1]} \odot A^{[1]}(1-A^{[1]})
$$

Then:

$$
dW^{[1]} = \frac{1}{m}X(dZ^{[1]})^T
$$

$$
db^{[1]} = \frac{1}{m}\sum_{\text{columns}} dZ^{[1]}
$$

# Update Rules

After computing the gradients, update parameters using gradient descent:

$$
W^{[2]} := W^{[2]} - \alpha dW^{[2]}
$$

$$
b^{[2]} := b^{[2]} - \alpha db^{[2]}
$$

$$
W^{[1]} := W^{[1]} - \alpha dW^{[1]}
$$

$$
b^{[1]} := b^{[1]} - \alpha db^{[1]}
$$

The most important idea is:

$$
\text{forward propagation computes predictions}
$$

while:

$$
\text{backpropagation computes how each parameter affected the prediction error}
$$

Then gradient descent uses those derivatives to reduce the loss.

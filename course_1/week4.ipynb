* Logistic regression is a shallow model. We get deeper neural networks as we increase hidden layers.

* Here we have a 4 layer network with 3 hidden layers so L=4 and $n^{[0]}=3$, $n^{[1]}=n^{[2]}=5$, $n^{[3]}=3$ and $n^{[4]}=1$.

* Forward Propagation means start with the input features, compute the activations of the first hidden layer, then use those activations to compute the next layer, and continue until the final output $\hat{y}$.

$$ \text{Let’s call the input layer } a^{[0]} = x$$

$$
a^{[0]} =
\begin{bmatrix}
x_1 \\
x_2 \\
x_3
\end{bmatrix}
$$

$$ \text{For the first hidden layer, each neuron computes two things, linear combination and activation.}$$


$$
z^{[1]} = W^{[1]}a^{[0]} + b^{[1]}
$$

$$
a^{[1]} = g^{[1]}(z^{[1]})
$$

$$\text{Here:}$$

$$W^{[1]} \quad \text{contains the weights from the input layer to the first hidden layer.}$$
$$b^{[1]} \quad \text{is the bias vector for the first hidden layer.}$$
$$z^{[1]} \quad \text{is the pre-activation value.}$$
$$a^{[1]} \quad \text{is the activation output of that layer.}$$
$$g^{[1]} \quad \text{might be ReLU, sigmoid, tanh, etc.}$$

$$\text{For the next hidden layer, the process is exactly the same:}$$

$$
z^{[2]} = W^{[2]}a^{[1]} + b^{[2]}
$$

$$
a^{[2]} = g^{[2]}(z^{[2]})
$$

$$\text{Then again:}$$

$$
z^{[3]} = W^{[3]}a^{[2]} + b^{[3]}
$$

$$
a^{[3]} = g^{[3]}(z^{[3]})
$$

$$\text{And so on, until the output layer.}$$

$$\text{In General For any layer } l$$

$$
z^{[l]} = W^{[l]}a^{[l-1]} + b^{[l]}
$$

$$
a^{[l]} = g^{[l]}(z^{[l]})
$$

* For the Network in the picture the input is


$$
a^{[0]} = x
$$

$$\text{The first hidden layer receives } x_1, x_2, x_3 \text{ and produces activations:}$$

$$
a^{[1]}
$$

$$\text{The second hidden layer takes } a^{[1]} \text{ as input and produces:}$$

$$
a^{[2]}
$$

$$\text{The third hidden layer takes } a^{[2]} \text{ as input and produces:}$$

$$
a^{[3]}
$$

$$\text{Then there appears to be another smaller hidden layer before the final output:}$$

$$
a^{[4]}
$$

$$\text{Finally, the output layer computes:}$$

$$
z^{[5]} = W^{[5]}a^{[4]} + b^{[5]}
$$

$$
\hat{y} = a^{[5]} = g^{[5]}(z^{[5]})
$$

$$\text{So the final prediction is:}$$

$$
\hat{y}
$$

* Intuition is each layer learns a different representation of the input. Early layers might learn simple combinations of $x_1, x_2, x_3$. Later layers combine those earlier patterns into more complex patterns. By the time we get to the final layer, the network has transformed the original input into something useful for making a prediction.

$$\text{So forward propagation is just the process of passing information from left to right:}$$

$$
x \rightarrow a^{[1]} \rightarrow a^{[2]} \rightarrow a^{[3]} \rightarrow a^{[4]} \rightarrow \hat{y}
$$

$$\text{You can summarize the whole thing as:}$$

$$
a^{[0]} = x
$$

$$\text{For } l = 1, 2, \dots, L$$

$$
z^{[l]} = W^{[l]}a^{[l-1]} + b^{[l]}
$$

$$
a^{[l]} = g^{[l]}(z^{[l]})
$$

$$\text{And at the final layer:}$$

$$
a^{[L]} = \hat{y}
$$

* Deep networks can learn hierarchical representations, where early layers learn simple features and later layers combine those simple features into more complex ones. For face recognition or face detection, a deep neural network might learn features in stages. The input is an image of a face. In the first few layers, the network may detect simple features such as edges. For example, one hidden unit might detect vertical edges, while another hidden unit might detect horizontal edges. 

<img src="./images/dl_17.png" width=800 height=300 />


* Finally, even deeper layers combine these face parts to detect or recognize complete faces. So the overall hierarchy is below. This is called a hierarchical representation or compositional representation. The earlier layers learn relatively simple functions of the input. Later layers combine those simple features into more complex features. So deep learning works well partly because it can build complex representations step by step.

$$
\text{Pixels} \rightarrow \text{Edges} \rightarrow \text{Face parts} \rightarrow \text{Faces}
$$

* The same idea applies to speech recognition. The input is an audio clip.

$$
\text{Audio waveform} \rightarrow \text{low-level sounds} \rightarrow \text{phonemes} \rightarrow \text{words} \rightarrow \text{sentences}
$$

* There is an analogy between deep neural networks and the human brain. The brain may also process information hierarchically. For example, in vision we have the map below. However, comparisons between deep learning and the human brain can be dangerous if taken too literally. Still, the idea of building simple features into complex features has been a loose inspiration for deep learning.

$$
\text{Simple visual patterns} \rightarrow \text{edges} \rightarrow \text{objects} \rightarrow \text{faces}
$$

* Another explanation for why deep networks work well comes from circuit theory. Circuit theory studies what kinds of functions can be computed using logic gates such as AND, OR, NOT. The important idea is that some functions can be computed efficiently by a deep network, but would require exponentially many units in a shallow network. A XOR Example. Suppose we want to compute the XOR, or parity, of many input features:

$$
y = x_1 \oplus x_2 \oplus x_3 \oplus \cdots \oplus x_n
$$

$$\text{A deep network can compute this efficiently using an XOR tree. First, it computes XORs of pairs:}$$

$$
x_1 \oplus x_2
$$

$$
x_3 \oplus x_4
$$

$$\text{Then it combines those results:}$$

$$
(x_1 \oplus x_2) \oplus (x_3 \oplus x_4)
$$

$$\text{Continuing this process creates a tree-like computation. The depth needed is approximately:}$$

$$
O(\log n)
$$

$$\text{So a deep network can compute this function with relatively few units.}$$

* Shallow Network Problem - If we force the network to have only one hidden layer, then computing the same XOR/parity function becomes much harder. A shallow network may need to enumerate many possible input configurations. There are $2^n$ possible configurations of n binary input features. So a shallow network may require on the order of $ O(2^n)$ hidden units. This means the shallow network could need exponentially more hidden units than a deep network.

* Deep vs. Shallow Networks - The key comparison is:

$$
\text{Deep network size} \approx O(\log n)
$$

$$\text{versus:}$$

$$
\text{Shallow network size} \approx O(2^n)
$$

$$\text{This shows that some functions are much easier to represent with deep networks than with shallow ones.}$$

* Practical Advice - Even though deep networks can work very well, you do not always need many layers. When starting a new problem, it can be useful to try simpler models first, such as Logistic regression. Then try a neural network with one or two hidden layers. The number of hidden layers is a hyperparameter that you can tune. So the depth of the network should be chosen based on performance, not just because deeper sounds better.

$$

$$

$$

$$

$$

$$\text{and so on. These gradients are what allow the network to update its parameters.}$$

* Once backpropagation has computed the gradients, the parameters are updated using gradient descent. For each layer:

$$
W^{[l]} := W^{[l]} - \alpha dW^{[l]}
$$

$$
b^{[l]} := b^{[l]} - \alpha db^{[l]}
$$

$$\text{where } \alpha \text{ is the learning rate. This completes one iteration of training.}$$


* The cache is used to pass information from the forward function to the backward function. Conceptually, the cache stores:

$$
Z^{[l]}
$$

$$\text{In implementation, the cache may also store:}$$

$$
W^{[l]}
$$


$$
b^{[l]}
$$

$$\text{This is a practical programming convenience because the backward function needs access to these values when computing gradients.}$$

* A deep neural network is built by repeating two main building blocks for each layer - Forward propagation, which computes activations and Backward propagation, which computes gradients. Each layer stores a cache during forward propagation so that backpropagation can efficiently compute derivatives later. One full training iteration consists of:

$$
X \rightarrow \text{Forward Propagation} \rightarrow \hat{Y}
$$

$$
\hat{Y} \rightarrow \text{Backpropagation} \rightarrow dW, db \rightarrow \text{ parameter updates}
$$

# Implementing Forward Propagation and Backpropagation

* The main idea is that each layer has a forward step that computes activations, and a corresponding backward step that computes gradients.

* Forward Propagation - For layer $l$, the forward propagation step takes the activation from the previous layer:

$$
A^{[l-1]}
$$

$$\text{and outputs the activation for the current layer:}$$

$$
A^{[l]}
$$

$$\text{It also stores values in a cache, such as:}$$

$$
Z^{[l]}, W^{[l]}, b^{[l]}
$$

$$\text{These cached values are later used during backpropagation. The forward propagation equations are:}$$

$$
Z^{[l]} = W^{[l]}A^{[l-1]} + b^{[l]}
$$

$$
A^{[l]} = g^{[l]}(Z^{[l]})
$$


$$W^{[l]} \text{ is the weight matrix for layer } l $$
$$b^{[l]} \text{ is the bias vector for layer } l $$
$$Z^{[l]} \text{ is the linear part of the computation}  $$
$$A^{[l]} \text{ is the activation output for layer } l $$
$$g^{[l]} \text{ is the activation function for layer } l $$


* Initializing Forward Propagation Forward propagation starts with the input data:

$$
A^{[0]} = X
$$

$$\text{Then the network repeatedly applies the forward propagation step from left to right:}$$

$$
A^{[0]} \rightarrow A^{[1]} \rightarrow A^{[2]} \rightarrow \cdots \rightarrow A^{[L]}
$$

$$\text{The final activation is the prediction:}$$

$$
A^{[L]} = \hat{Y}
$$

* Vectorized Forward Propagation - In a vectorized implementation, the same equations are used, but they are applied to all training examples at once. The vectorized equation is:

$$
Z^{[l]} = W^{[l]}A^{[l-1]} + b^{[l]}
$$

$$
A^{[l]} = g^{[l]}(Z^{[l]})
$$

$$\text{The bias term } b^{[l]} \text{ is added using Python broadcasting. If there are m training examples, then:}$$

$$
A^{[l-1]} : n^{[l-1]} \times m
$$

$$
W^{[l]} : n^{[l]} \times n^{[l-1]}
$$

$$
b^{[l]} : n^{[l]} \times 1
$$

$$
Z^{[l]} : n^{[l]} \times m
$$

$$
A^{[l]} : n^{[l]} \times m
$$

* Backward propagation computes the derivatives needed to update the parameters. For layer $l$, the backward step takes:

$$
dA^{[l]}
$$

$$\text{and outputs:}$$

$$
dA^{[l-1]}
$$

$$
dW^{[l]}
$$

$$
db^{[l]}
$$

$$\text{The first step is to compute:}$$

$$
dZ^{[l]} = dA^{[l]} * g^{[l]'}(Z^{[l]})
$$

$$\text{where } * \text{ means element-wise multiplication. Then compute the parameter gradients:}$$

$$
dW^{[l]} = dZ^{[l]}A^{[l-1]T}
$$

$$
db^{[l]} = dZ^{[l]}
$$

$$\text{Finally, compute the gradient passed to the previous layer:}$$

$$
dA^{[l-1]} = W^{[l]T}dZ^{[l]}
$$

* Vectorized Backward Propagation - For a vectorized implementation with m training examples, the equations become:

$$
dZ^{[l]} = dA^{[l]} * g^{[l]'}(Z^{[l]})
$$

$$
dW^{[l]} = \frac{1}{m}dZ^{[l]}A^{[l-1]T}
$$

$$
db^{[l]} = \frac{1}{m}\text{np.sum}(dZ^{[l]}, \text{axis}=1, \text{keepdims}=\text{True})
$$

$$
dA^{[l-1]} = W^{[l]T}dZ^{[l]}
$$

$$\text{The factor appears because the gradients are averaged over all training examples:}$$

$$
\frac{1}{m}
$$


* Three-Layer Network Example - For a three-layer neural network, the forward propagation might look like this:

$$
X \rightarrow \text{ReLU} \rightarrow \text{ReLU} \rightarrow \text{Sigmoid} \rightarrow \hat{Y}
$$

$$\text{For example:}$$

$$
A^{[0]} = X
$$

$$
Z^{[1]} = W^{[1]}A^{[0]} + b^{[1]}
$$

$$
A^{[1]} = \text{ReLU}(Z^{[1]})
$$

$$
Z^{[2]} = W^{[2]}A^{[1]} + b^{[2]}
$$

$$
A^{[2]} = \text{ReLU}(Z^{[2]})
$$

$$
Z^{[3]} = W^{[3]}A^{[2]} + b^{[3]}
$$

$$
A^{[3]} = \sigma(Z^{[3]})
$$

$$
A^{[3]} = \hat{Y}
$$

$$\text{If this is binary classification, the final activation is often sigmoid:}$$

$$
\sigma(z) = \frac{1}{1 + e^{-z}}
$$

* Computing the Loss - After forward propagation, the prediction $\hat{Y}$ is used to compute the loss. For binary classification, the loss for one example is usually:

$$
\mathcal{L}(\hat{y}, y) = -y\log(\hat{y}) - (1-y)\log(1-\hat{y})
$$

$$\text{The loss tells us how far the prediction is from the true label.}$$

* Starting Backpropagation - Backward propagation starts by computing the derivative of the loss with respect to the final activation:

$$
dA^{[L]}
$$

$$\text{For binary classification with logistic regression, this derivative is:}$$

$$
dA^{[L]} = -\frac{Y}{A^{[L]}} + \frac{1-Y}{1-A^{[L]}}
$$

$$\text{Equivalently:}$$

$$
dA^{[L]} = \frac{1-Y}{1-A^{[L]}} - \frac{Y}{A^{[L]}}
$$

$$\text{This initializes the backward pass. Then the network works backward through the layers:}$$

$$
dA^{[L]} \rightarrow dA^{[L-1]} \rightarrow dA^{[L-2]} \rightarrow \cdots \rightarrow dA^{[0]}
$$

$$\text{Along the way, it computes:}$$

$$
dW^{[L]}, db^{[L]}
$$

$$
dW^{[L-1]}, db^{[L-1]}
$$

$$
\cdots
$$

$$\text{The value } dA^{[0]} \text{ may be computed, but it is usually not needed because } A^{[0]} \text{ is just the input data X}$$

$$
dW^{[1]}, db^{[1]}
$$



* Role of the Cache - During forward propagation, the network stores intermediate values in a cache. For example:

$$
Z^{[1]}, Z^{[2]}, Z^{[3]}
$$

$$\text{and often:}$$

$$
W^{[l]}, b^{[l]}, A^{[l-1]}
$$

$$\text{These cached values are needed during backpropagation. The forward pass computes predictions, while the backward pass uses cached values to compute gradients.}$$

* Full Training Flow

$$
X \rightarrow \text{Forward Propagation} \rightarrow \hat{Y}
$$

$$
\hat{Y} \rightarrow \text{Loss}
$$


$$
\text{Loss} \rightarrow \text{Backward Propagation} \rightarrow dW, db
$$


$$
W^{[l]} := W^{[l]} - \alpha dW^{[l]}
$$

$$
b^{[l]} := b^{[l]} - \alpha db^{[l]}
$$



* Key Takeaways - Forward propagation computes predictions. Backward propagation computes gradients and moves from the loss back through the network

$$
Z^{[l]} = W^{[l]}A^{[l-1]} + b^{[l]}
$$

$$
A^{[l]} = g^{[l]}(Z^{[l]})
$$


$$
dZ^{[l]} = dA^{[l]} * g^{[l]'}(Z^{[l]})
$$

$$
dW^{[l]} = \frac{1}{m}dZ^{[l]}A^{[l-1]T}
$$

$$
db^{[l]} = \frac{1}{m}\text{np.sum}(dZ^{[l]}, \text{axis}=1, \text{keepdims}=\text{True})
$$

$$
dA^{[l-1]} = W^{[l]T}dZ^{[l]}
$$


$$
X \rightarrow \hat{Y}
$$


$$
dA^{[L]} \rightarrow dA^{[L-1]} \rightarrow \cdots \rightarrow dA^{[0]}
$$




# Parameters and Hyperparameters in Deep Learning

* The parameters of a neural network are the values the model learns during training. The main parameters are W and b. These are learned by the training algorithm, usually through gradient descent.

* Hyperparameters are values that you choose before or during training. They control how the learning algorithm behaves. Examples of hyperparameters include learning rate $\alpha$, the number of gradient descent steps, number of layers L, the number of hidden units in each layer, and the choice of activation function, such as ReLU, tanh or sigmoid. All these have an impact on how W and b change during training. All the hyperparameters require trying different values to see which work the best. Good hyperparameters for one application may not work well for another. 

* Why Learning Rate Is a Hyperparameter? It controls how large each gradient descent update is. If too small, learning may be very slow. If too large, the cost function may fail to decrease or may even diverge. So the learning rate affects how the parameters evolve during training. So you experiment with different values until you find one that makes the cost decrease quickly and reliably. That is why $\alpha$ is called a hyperparameter rather than a regular parameter.

* Other Hyperparameters (momentum term, mini-batch size, regularization parameters) we will get to later but the key point is that deep learning involves many choices that affect training.

* To choose hyperparameters, you should evaluate performance on a holdout validation set. The training set is used to learn W and b. The validation set is used to choose hyperparameters.

# What Deep Learning Has to Do with the Brain

* A deep learning neuron is a highly simplified mathematical model of a biological neuron. While both process information by passing inputs through weighted connections to generate an output, they differ fundamentally in structure, complexity, and how they learn.
* A biological neuron is a complex, living cell that relies on electrochemical signals and physical structures to function.
    * Structure: Composed of dendrites (which receive electrical signals from other neurons), the soma (cell body that processes signals), and the axon (which transmits signals to other neurons).
    * Signaling: Communicates using discrete electrical impulses called "action potentials" or spikes.
    * Learning & Adaptation: Continuously adapts via synaptic plasticity, where the connections physically strengthen or weaken based on experience (e.g., Hebbian learning).
    * Complexity: A single neuron is capable of intricate, non-linear computations on its own.

<img src="./images/dl_20.png" width=900 height=500 />

* A deep learning neuron (also called a node or perceptron) is a mathematical function designed to crudely mimic biological information processing.
    * Structure: It receives numerical inputs, multiplies them by assigned weights (representing synaptic strength), adds a bias, and passes the result through an activation function (e.g., ReLU or Sigmoid) to generate an output.
    * Signaling: Uses continuous, floating-point numbers (such as activations) rather than binary, all-or-nothing electrical spikes.
    * Learning & Adaptation: Learns by adjusting weights across an entire network using algorithms like backpropagation and gradient descent, which require massive amounts of structured training data.
    * Complexity: A single artificial neuron is computationally simple; intelligence emerges from layering millions of them together.
* Deep learning networks were directly inspired by the architecture of the human brain. Both systems process information hierarchically—routing raw inputs through layers to recognize complex patterns (like identifying an edge or an entire object). They share conceptual similarities in how experience improves output, leading to the use of terms like "neural networks" and "synapses".
* However, modern neuroscientists widely agree that the analogy is largely metaphorical. While deep learning models can perform specific cognitive tasks with high accuracy, they do not accurately reflect the physical, chemical, and temporal complexities of actual brains.

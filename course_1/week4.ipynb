# Deep Neural Networks

* Logistic regression is a shallow model. We get deeper neural networks as we increase hidden layers.

* Here we have a 4 layer network with 3 hidden layers so L=4 and $n^{[0]}=3$, $n^{[1]}=n^{[2]}=5$, $n^{[3]}=3$ and $n^{[4]}=1$. 

<img src="./images/dl_15.png" width=800 height=300 />

* Forward Propagation means start with the input features, compute the activations of the first hidden layer, then use those activations to compute the next layer, and continue until the final output $\hat{y}$.

$$ \text{Letâ€™s call the input layer } a^{[0]} = x$$

$$
a^{[0]} =
\begin{bmatrix}
x_1 \\
x_2 \\
x_3
\end{bmatrix}
$$

$$ \text{For the first hidden layer, each neuron computes two things, linear combination and activation.}$$


$$
z^{[1]} = W^{[1]}a^{[0]} + b^{[1]}
$$

$$
a^{[1]} = g^{[1]}(z^{[1]})
$$

$$\text{Here:}$$

$$W^{[1]} \quad \text{contains the weights from the input layer to the first hidden layer.}$$
$$b^{[1]} \quad \text{is the bias vector for the first hidden layer.}$$
$$z^{[1]} \quad \text{is the pre-activation value.}$$
$$a^{[1]} \quad \text{is the activation output of that layer.}$$
$$g^{[1]} \quad \text{might be ReLU, sigmoid, tanh, etc.}$$

$$\text{For the next hidden layer, the process is exactly the same:}$$

$$
z^{[2]} = W^{[2]}a^{[1]} + b^{[2]}
$$

$$
a^{[2]} = g^{[2]}(z^{[2]})
$$

$$\text{Then again:}$$

$$
z^{[3]} = W^{[3]}a^{[2]} + b^{[3]}
$$

$$
a^{[3]} = g^{[3]}(z^{[3]})
$$

$$\text{And so on, until the output layer.}$$

$$\text{In General For any layer } l$$

$$
z^{[l]} = W^{[l]}a^{[l-1]} + b^{[l]}
$$

$$
a^{[l]} = g^{[l]}(z^{[l]})
$$

* For the Network in the picture the input is


$$
a^{[0]} = x
$$

$$\text{The first hidden layer receives } x_1, x_2, x_3 \text{ and produces activations:}$$

$$
a^{[1]}
$$

$$\text{The second hidden layer takes } a^{[1]} \text{ as input and produces:}$$

$$
a^{[2]}
$$

$$\text{The third hidden layer takes } a^{[2]} \text{ as input and produces:}$$

$$
a^{[3]}
$$

$$\text{Then there appears to be another smaller hidden layer before the final output:}$$

$$
a^{[4]}
$$

$$\text{Finally, the output layer computes:}$$

$$
z^{[5]} = W^{[5]}a^{[4]} + b^{[5]}
$$

$$
\hat{y} = a^{[5]} = g^{[5]}(z^{[5]})
$$

$$\text{So the final prediction is:}$$

$$
\hat{y}
$$

* Intuition is each layer learns a different representation of the input. Early layers might learn simple combinations of $x_1, x_2, x_3$. Later layers combine those earlier patterns into more complex patterns. By the time we get to the final layer, the network has transformed the original input into something useful for making a prediction.

$$\text{So forward propagation is just the process of passing information from left to right:}$$

$$
x \rightarrow a^{[1]} \rightarrow a^{[2]} \rightarrow a^{[3]} \rightarrow a^{[4]} \rightarrow \hat{y}
$$

$$\text{You can summarize the whole thing as:}$$

$$
a^{[0]} = x
$$

$$\text{For } l = 1, 2, \dots, L$$

$$
z^{[l]} = W^{[l]}a^{[l-1]} + b^{[l]}
$$

$$
a^{[l]} = g^{[l]}(z^{[l]})
$$

$$\text{And at the final layer:}$$

$$
a^{[L]} = \hat{y}
$$

$$\text{the weight matrices have these dimensions}$$

$$
W^{[1]} : 3 \times 2
$$

$$
W^{[2]} : 5 \times 3
$$

$$
W^{[3]} : 4 \times 5
$$

$$
W^{[4]} : 2 \times 4
$$

$$
W^{[5]} : 1 \times 2
$$

$$\text{The general rule is:}$$

$$
W^{[l]} : n^{[l]} \times n^{[l-1]}
$$


$$
\text{rows} = \text{number of units in the current layer}
$$

$$
\text{columns} = \text{number of units in the previous layer}
$$

$$\text{the bias matrices have these dimensions}$$


$$
b^{[1]} : 3 \times 1
$$

$$
b^{[2]} : 5 \times 1
$$

$$
b^{[3]} : 4 \times 1
$$

$$
b^{[4]} : 2 \times 1
$$

$$
b^{[5]} : 1 \times 1
$$

$$\text{The general rule is:}$$

$$
b^{[l]} : n^{[l]} \times 1
$$

$$\text{Dimensions of Activations and Pre-Activations:}$$


$$
z^{[l]} : n^{[l]} \times 1
$$

$$
a^{[l]} : n^{[l]} \times 1
$$

$$\text{Since the activation function is applied element-wise, } z^{[l]} \text{ and } a^{[l]} \text{ have the same shape.}$$

$$\text{For backpropagation, the gradients must match the dimensions of the parameters.}$$



$$
dW^{[l]} \text{ has the same dimension as } W^{[l]}
$$


$$
db^{[l]} \text{ has the same dimension as } b^{[l]}
$$


$$
dW^{[l]} : n^{[l]} \times n^{[l-1]}
$$

$$
db^{[l]} : n^{[l]} \times 1
$$


$$
dZ^{[l]} \text{ has the same dimension as } Z^{[l]}
$$


$$
dA^{[l]} \text{ has the same dimension as } A^{[l]}
$$

* Vectorized Implementation with m examples


$$
X = A^{[0]}
$$

$$\text{Instead of a single example with shape:}$$

$$
n^{[0]} \times 1
$$


$$\text{we stack all m examples horizontally, giving:}$$

$$
X = A^{[0]} : n^{[0]} \times m
$$

$$\text{For each layer:}$$

$$
Z^{[l]} : n^{[l]} \times m
$$

$$
A^{[l]} : n^{[l]} \times m
$$

$$\text{The vectorized forward propagation equation is:}$$

$$
Z^{[l]} = W^{[l]}A^{[l-1]} + b^{[l]}
$$

$$
W^{[l]} : n^{[l]} \times n^{[l-1]}
$$

$$
A^{[l-1]} : n^{[l-1]} \times m
$$

$$
Z^{[l]} : n^{[l]} \times m
$$

$$
b^{[l]} : n^{[l]} \times 1
$$

* Through Python broadcasting, $b^{[l]}$ is copied across the m columns and added element-wise. Keeping track of dimensions is one of the most useful debugging techniques when implementing deep neural networks. The most important rules are:

$$
W^{[l]} : n^{[l]} \times n^{[l-1]}
$$

$$
b^{[l]} : n^{[l]} \times 1
$$

$$
Z^{[l]}, A^{[l]} : n^{[l]} \times m
$$

$$
dW^{[l]} = \text{same shape as } W^{[l]}
$$

$$
db^{[l]} = \text{same shape as } b^{[l]}
$$

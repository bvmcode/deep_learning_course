# Neural Network Basics

* A neural network is a system that learns how to combine inputs in increasingly complex ways to make accurate predictions, like estimating house prices from multiple interacting factors.
* Think of a neural network as a flexible function that learns how to map inputs → outputs by spotting patterns in data.
* We want to predict a house price (y) based on features seen below.

<img src="./images/dl_1.png" width="600" height="300">

* Instead of trying to write a formula ourselves, the neural network learns the formula automatically. It does this in layers.
    * Input layer - raw features
    * Hidden layers - Each neuron takes all inputes and multiplies by a weight, adds them up and passes the result through a non linear activation function
        * Each neuron learns a mini rule like 
            * Big houses in rich areas → higher price
            * More bedrooms only matters if size is large
            * Zip code changes everything
    * Output layer - The final neuron combines all those learned patterns and outputs predicted price.
* The intuition is
    * inputs = facts
    * hidden neurons = pattern detectors
    * output = final price
* Training a neural network is just repeating this loop
    * Forward propagation → make a prediction
    * Backpropagation → figure out how wrong it was and fix it
* Backpropagation, based on errors, figures out which connections (weights) contributed most to the wrong prediction and whether each weight should go up or down.
* In training the weights get slightly tweaked thousands of times.
* Some NNs are supervised like the above real estate example but some are unsupervised.
* They can use structured data (tables of data) and unstructured data (images, free text, audio).
* Types of Neural Networks

| Neural Network Type                      | What It’s Good At                   | How It Works (Intuition)                          | Common Use Cases                                            |
| ---------------------------------------- | ----------------------------------- | ------------------------------------------------- | ----------------------------------------------------------- |

| **Graph Neural Network (GNN)**           | Graph-structured data               | Learns relationships between nodes                | Social networks, fraud detection, molecules                 |

# Logistic Regression

* Logistic Regression is a **supervised learning algorithm** used for **binary classification**. It predicts the probability that an input belongs to a particular class (e.g., 0 or 1).
* $m$ = number of training examples (samples)
* $n$ = number of features (dimensions)
* Input data matrix:
  $$
  X \in \mathbb{R}^{m \times n}
  $$
* Labels:
  $$
  y \in \mathbb{R}^{m \times 1}
  $$
* Weights:
  $$
  w \in \mathbb{R}^{n \times 1}
  $$
* Bias:
  $$
  b \in \mathbb{R}
  $$

* Summary of Shapes

| Variable  | Shape    |
| --------- | -------- |
| $X$       | $(m, n)$ |
| $x^{(i)}$ | $(n,)$   |
| $w$       | $(n, 1)$ |
| $b$       | scalar   |
| $Z$       | $(m, 1)$ |
| $\hat{Y}$ | $(m, 1)$ |

* Key Intuition
    * Logistic regression is a **linear model + nonlinear activation (sigmoid)**
    * It maps inputs to **probabilities**
    * Works well when classes are **linearly separable**
    * Scales to high-dimensional inputs like images (e.g., 12288 features)

* In logistic regression, weights and the intercept (bias) are the parameters the model learns to make predictions. Weights $w$ determines the orientation (slope) of the boundary and intercept $b$ determines the position (shift) of the boundary.
* Weights are the coefficients applied to each feature. Each weight tells you:
    * How important a feature is
    * How strongly it influences the prediction
    * Whether it pushes the prediction toward 0 or 1
* For example
    * $w_1 = 2$ → strong positive influence
    * $w_2 = -1$ → pushes prediction down
    * $w_3 = 0.5$ → mild positive influence
* The intercept is a constant added to the model. It shifts the decision boundary left or right. It represents the model’s output when all features are zero

* Consider a 64x64 image (Red, Green, Blue). Each channel becomes a vector of length $64 \times 64 = 4096$. We stack the channels so that a single input vector is of dimension $\text{Total features} = 3 \times 4096 = 12288$. In our input matrix X each row is one flattened image.

<img src="./images/dl_2.png" width="600" height="300">

* Each pixel (R, G, B value) has its own weight. The model learns which pixels matter, which colors matter and which regions of the image are important.

* Logistic regression computes:

    $$
    z = w^T x + b
    $$

* Then applies the sigmoid function:

    $$
    \sigma(z) = \frac{1}{1 + e^{-z}}
    $$

* So the prediction is:

    $$
    \hat{y} = \sigma(w^T x + b)
    $$

* Vectorized Form (All Samples)

    $$
    Z = X w + b
    $$

  $$\text{X: } (m \times n)$$
  $$\text{w: } (n \times 1)$$
  $$\text{Z: } (m \times 1)$$

    $$
    \hat{Y} = \sigma(Z)
    $$

* Decision Boundary to get class

    $$
    \hat{y} =
    \begin{cases}
    1 & \text{if } \sigma(z) \geq 0.5 \
    0 & \text{otherwise}
    \end{cases}
    $$

    $$
    w^T x + b \geq 0
    $$

* Loss Function (Binary Cross-Entropy) is smooth and convex allowing for gradient descent optimization. Binary cross-entropy is used because it naturally arises from probability theory and gives the right learning behavior for classification problems.

    $$\text{For one example:}$$

    $$
    \mathcal{L}(\hat{y}, y) = - \left[ y \log(\hat{y}) + (1 - y)\log(1 - \hat{y}) \right]
    $$

    $$\text{For all examples:}$$

    $$
    J(w, b) = \frac{1}{m} \sum_{i=1}^{m} \mathcal{L}(\hat{y}^{(i)}, y^{(i)})
    $$

* Here y is a class label and y-hat is a probability.
* If y=1 then loss becomes $\mathcal{L}=-\log(\hat{y})$. Only $\hat{y}$ matters here.
    * If $\hat{y} \to 1$ → $\log(\hat{y}) \to 0$ → loss is small.
    * If $\hat{y} \to 0$ → $\log(\hat{y}) \to -\infty$ → loss is huge

* If y=0 then loss becomes $\mathcal{L}=-\log(1-\hat{y})$. Only $\hat{y}$ matters here.
    * If $\hat{y} \to 0$ → $\log(1 - \hat{y}) \to 0$ → loss is small
    * If $\hat{y} \to 1$ → $\log(1 - \hat{y}) \to -\infty$ → loss is huge

* Gradient descent is how the model learns the right weights ($w$) and bias ($b$). We update parameters to minimize cost where $\alpha$ is the learning rate. It repeatedly asks: “How should I tweak my parameters to reduce my mistakes?” The gradient tells you which direction increases the loss the most so if you go in the opposite direction, you reduce the loss. In the image example its learning which pixels matter and in what direction. 

    $$
    \frac{\partial J}{\partial w} = \frac{1}{m} X^T (\hat{Y} - Y)
    $$

    $$
    \frac{\partial J}{\partial b} = \frac{1}{m} \sum_{i=1}^{m} (\hat{y}^{(i)} - y^{(i)})
    $$

    $$
    w := w - \alpha \frac{\partial J}{\partial w}
    $$

    $$
    b := b - \alpha \frac{\partial J}{\partial b}
    $$

* A computational graph in deep learning is a directed graph representing mathematical operations, where nodes are variables or operations and edges are data dependencies. It enables automatic differentiation (backpropagation) to compute gradients efficiently by breaking complex functions into smaller, chained operations. Frameworks like TensorFlow and PyTorch use this structure to update model weights.

* Neural networks compute outputs in a forward pass and compute gradients (derivatives) in a backward pass. A computation graph breaks a function into simple step-by-step operations. Say in this example A=5, B=3 and C=2.

<img src="./images/dl_5.png" width="600" height="300">

* So instead of one big calculation, it’s broken into smaller parts. The computation graph shows how variables flow through these steps. A forward pass (left → right) computes the final output J. A backward pass (right → left) computes derivatives needed for learning.

* This structure makes it easy to compute outputs efficiently and apply backpropagation to find gradients for optimization (like minimizing a cost function). Computation graphs organize calculations so neural networks can compute outputs and learn from errors efficiently.

* Now let's see how backpropagation works
    * First step of backprop -> $dJ/dV=3$
    * Chain rule - $dJ/dA = dJ/dV \cdot dV/dA$
        * since V = A + U -> $dV/dA = 1$
        * $dJ/dA = dJ/dV = 3$
    * Continue backward
        * $dJ/dU =3$
        * $dJ/dB = dJ/dU \cdot dU/dB =3 \cdot C=6$
            * If B increases slightly, J increases 6× as much
            * So gradients tell you which inputs matter most
        * $dJ/dC = dJ/dU \cdot dU/dC=3 \cdot B=9$

* Instead of writing full derivatives:
    * dV means dJ/dV
    * dA means dJ/dA
    * and so on

* So backprop = chain rule + reuse. Backpropagation is just the chain rule applied efficiently. The big win is reusing intermediate gradients, avoiding recomputation.
* Right-to-left is not arbitrary. Forward pass computes values and backward pass computes sensitivities (how much each variable affects output).
* Efficiency is everything. Naively computing derivatives separately is expensive. Backprop reduces complexity from exponential → linear in graph size.
* Each node only needs to know its input and compute output on forward pass and how to compute its local derivative on backprop with respect to the edges coming into it.
* Here is another example

$$
L(a,b,c) = c(a + 2b)
$$

* We decompose it into intermediate steps

$$
d = 2b
$$
$$
e = a + d
$$
$$
L = c \cdot e
$$

* This forms a graph where:
    * Inputs: $a=3, b=1, c=-2$
    * Intermediate nodes: $d, e$
    * Output: $L$

* Foward pass - Compute values from inputs → output:

$$
d = 2b = 2(1) = 2
$$
$$
e = a + d = 3 + 2 = 5
$$
$$
L = c \cdot e = (-2)(5) = -10
$$

<img src="./images/dl_6.png" width="600" height="300">

* Backpropagation computes gradients using the **chain rule**, moving from output → inputs.

$$\text{Step 1: Start from output}$$

$$
\frac{\partial L}{\partial L} = 1
$$

$$\text{Step 2: Gradients at final node } L = ce$$


$$
\frac{\partial L}{\partial c} = e = 5
$$
$$
\frac{\partial L}{\partial e} = c = -2
$$


$$\text{Step 3: Gradients at } e = a + d$$

$$
\frac{\partial e}{\partial a} = 1, \quad \frac{\partial e}{\partial d} = 1
$$

$$\text{Apply chain rule:}$$
$$
\frac{\partial L}{\partial a} = \frac{\partial L}{\partial e} \cdot \frac{\partial e}{\partial a} = (-2)(1) = -2
$$
$$
\frac{\partial L}{\partial d} = \frac{\partial L}{\partial e} \cdot \frac{\partial e}{\partial d} = (-2)(1) = -2
$$


$$\text{Step 4: Gradients at } d = 2b$$

$$
\frac{\partial d}{\partial b} = 2
$$

$$\text{Apply chain rule:}$$
$$
\frac{\partial L}{\partial b} = \frac{\partial L}{\partial d} \cdot \frac{\partial d}{\partial b} = (-2)(2) = -4
$$

$$\text{Final Gradients}$$

$$
\frac{\partial L}{\partial a} = -2, \quad
\frac{\partial L}{\partial b} = -4, \quad
\frac{\partial L}{\partial c} = 5
$$

<img src="./images/dl_8.png" width="600" height="300">

* The chain rule says: to see how a change propagates through a system, multiply the effect at each step. Each term is a **local sensitivity**, and multiplying them gives the **total sensitivity**.

$$
b \rightarrow d \rightarrow e \rightarrow L
$$

$$
\frac{\partial L}{\partial b}=

\frac{\partial L}{\partial e}
\cdot
\frac{\partial e}{\partial d}
\cdot
\frac{\partial d}{\partial b}
$$


*  Key Takeaways
    * A **computation graph** organizes calculations into reusable steps.
    * The **forward pass** computes values.
    * The **backward pass** computes gradients using the **chain rule**.
    * Gradients **flow backward**, and intermediate results are reused.
    * This is the foundation of training neural networks (gradient descent).

* The above was for one training example. Looks see how we can do it for all.

    $$\text{average loss}$$
    $$J(w,b) = \frac{1}{m} \sum_{i=1}^m \mathcal{L} (a^{(i)}, y^{(i)})$$

    $$\text{----------------------------------------------}$$

    $$\text{set J=dw=db=0 }$$
    $$\text{ AND do the below in for loop i=1 to m}$$

    $$a^{(i)} = \hat{y}^{(i)} = \sigma(z^{(i)}) = \sigma(w^T x^{(i)} +b)$$
    $$J +=  -\left[y^{(i)} \log a^{(i)} + (1 - y^{(i)})\log(1 - a^{(i)})\right]$$

    $$dz^{(i)} = \frac{\partial J}{\partial z^{(i)}} = a^{(i)} - y^{(i)}$$
    $$dw_1 = \frac{\partial J}{\partial w_1} += x_1^{(i)} dz^{(i)}$$
    $$dw_2 = \frac{\partial J}{\partial w_2} += x_2^{(i)} dz^{(i)}$$
    $$db = \frac{\partial J}{\partial b} += dz^{(i)}$$

    $$\text{----------------------------------------------}$$

    $$\text{after the loop}$$
    $$J /= m$$
    $$dw_1 /= m$$
    $$dw_2 /= m$$
    $$db /= m$$
 
    $$\text{----------------------------------------------}$$

    $$\text{then update}$$
    $$
    w_1 := w_1 - \alpha dw_1
    $$
    $$
    w_2 := w_2 - \alpha dw_2
    $$
    $$
    b := b - \alpha db
    $$

    $$\text{----------------------------------------------}$$

    $$\text{then do again till converge is done where } J=dw=db=0 \text{ again.}$$

# Vectorization

* Think of $z=w^T x + b$ for one training example. We would have to loop through the weights and inputs to calculate z.

```python
z=0
n=3
w=[.2, .3, .4]
x=[1.1, 1.2, .8]
b=2

for i in range(len(x)):
    z += w[i] * x[i]

z += b
print(z)
```

* Vectorization in numpy is the process of performing operations on entire arrays at once instead of using explicit "for-loops" to process one element at a time. Vectorized code can look a lot cleaner and be 10 to 100+ times faster than traditional Python loops.

```python
import numpy as np

z = np.dot(np.array(w), np.array(x)) + b
print(z)
```

* It makes code faster and cleaner by moving the heavy lifting from slow Python code to highly optimized machine-level instructions.
* Vectorized operations are executed in pre-compiled C or Fortran code. This code runs much closer to the computer's hardware than standard Python does.
* Modern CPUs have a feature called SIMD (Single Instruction, Multiple Data). Vectorization allows the CPU to load a "chunk" of numbers into special registers and perform the same math on all of them simultaneously in one clock cycle.
*  NumPy stores data in contiguous (touching) blocks of memory. This allows the CPU to grab large chunks of data quickly without jumping around different memory addresses, which is much faster. 
* Let's vectorize logistic regression. You have to go example by example in forward propagation.

$$z^{(i)} = w^Tx^{(i)}+b$$
$$a = \sigma(z^{(i)})$$

* But we can vectorize this across all training examples m. Recall X is a nXm matrix where n is the number of dimensions and m is the number of training examples. The weights w is a nX1 vector, intercepts b is a 1Xm vector all with the same value. So when we calculate z we end up with a 1Xm vector.

$$Z = w^T \cdot X + B$$

$$
Z = [
w_1 ,
w_2 ,
w_3 ,
\cdots 
w_n
]
\begin{bmatrix}
x^{(1)}_1 & x^{(2)}_1 & \cdots & x^{(m)}_1 \\
x^{(1)}_2 & x^{(2)}_2 & \cdots & x^{(m)}_2 \\
x^{(1)}_3 & x^{(2)}_3  & \cdots & x^{(m)}_3 \\ 
\vdots & \vdots & \ddots & \vdots \\ 
x^{(1)}_n & x^{(2)}_n & \cdots & x^{(m)}_n 
\end{bmatrix}

+ [b, b, \cdots b]$$
$$\text{Then we have a vector of length m made up of dot products plus scalar b.}$$
$$
= [w^T x^{(1)} +  b, w^T x^{(2)} +  b, \cdots, w^T x^{(m)} +  b]
$$
$$\text{Where we have a resulting vector of size } (1 \times n) (n \times m) = (1 \times m)$$
$$
= [(w_1 x^{(1)}_1 + \cdots + w_n x^{(1)}_n +  b), (w_1 x^{(2)}_1 + \cdots + w_n x^{(2)}_n +  b), \cdots, (w_1 x^{(m)}_1 + \cdots + w_n x^{(m)}_n +  b)]
$$

$$\text{And then we have a vector of probabilities of size } (1 \times m)$$
$$ A = \sigma(Z)$$


```python
import numpy as np
w = np.array([[.2],
              [.3],
              [.4]])
X = np.array([
    [1.1, 1.2, .8],
    [.5, .6, .7],
    [.9, .8, .3]
    ])
b = 2 # auto expand to np.array([2, 2, 2]) (broadcasting)
Z = np.dot(w.T,X) + b
print(Z)
```

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

A = sigmoid(Z)
print(A)
```

* Now let's add the gradients.

    $$A = [a^{(1)}, a^{(2)}, \cdots, a^{(m)},]$$
    $$Y = [y^{(1)}, y^{(2)}, \cdots, y^{(m)},]$$
    $$dZ = \frac{\partial J}{\partial Z} = [dz^{(1)}, dz^{(2)}, \cdots, dz^{(m)}] $$
    $$= A-Y = [a^{(1)}-y^{(1)}, a^{(2)}-y^{(2)}, \cdots, a^{(m)}-y^{(m)}]$$

* Recall we had the below across all m training examples.

    $$dw = \frac{\partial J}{\partial w} =  0$$
    $$dw += x^{(i)}dz^{(i)}$$
    $$dw /= m$$
    $$w := w - \alpha dw$$

    $$db = \frac{\partial J}{\partial b} =  0$$
    $$db += dz^{(i)}$$
    $$db /= m$$
    $$b := b - \alpha db$$

* We can vectorize these

    $$dw = \frac{1}{m} \sum_{i=1}^m x^{(i)}dz^{(i)}$$
    $$db = \frac{1}{m} \sum_{i=1}^m dz^{(i)}$$




```python
Y = np.array([[1, 0, 1]])
dZ = A - Y
dw = np.dot(X, dZ.T)/3
db = np.sum(dZ)/3
print(dw)
print(db)
```

```python
# putting it all together
import numpy as np

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def forward_propagation(X, w, b, Y, learn_rate=0.01, iterations=1000):
    for _ in range(iterations):
        Z = np.dot(w.T,X) + b
        A = sigmoid(Z)
        dZ = A - Y
        m = X.shape[1]
        dw = np.dot(X, dZ.T)/m
        db = np.sum(dZ)/m
        w = w - learn_rate * dw
        b = b - learn_rate * db
    return w, b


w = np.array([[.2],
              [.3],
              [.4]])
X = np.array([
    [1.1, 1.2, .8],
    [.5, .6, .7],
    [.9, .8, .3]
    ])
b = 2
Y = np.array([[1, 0, 1]])


w, b = forward_propagation(X, w, b, Y)
print(w)
print(b)
```
